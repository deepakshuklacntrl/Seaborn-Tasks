# Line Plot as scatter & facet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# line plot()

```python
sns.lineplot(data=tips,x='day',y='total_bill',hue='smoker',style='time')
plt.show()
```

# Scatter plot()

```python
sns.scatterplot(data=tips,x='day',y='total_bill',hue='smoker',style='time')
plt.show()
```

# Faceting(col)

```python
sns.relplot(data=tips,x='day',y='total_bill',col='sex',kind='line')
plt.show()
```
