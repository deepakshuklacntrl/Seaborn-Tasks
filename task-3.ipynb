# Distribution plot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# Histogram plot()

```python
sns.histplot(data=tips,x='total_bill',bins=15,hue='sex',element='step')
plt.show()
```

# Kernal Density Estimation(Kde plot)

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,hue='sex')
plt.show()
```

# Rug Plot()

```python
sns.rugplot(data=tips,x='total_bill',hue='sex')
plt.show()
```

# Hist + Kde Plot

```python
sns.histplot(data=tips,x='total_bill',bins=15,hue='sex',kde=True,element='step')
plt.show()
```
