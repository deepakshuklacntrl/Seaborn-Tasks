# Relational plot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Task-1 Relational Plot
1. rel plot

```python
sns.relplot(data=tips,x='day',y='total_bill')
plt.show()
```

# 2. Scatter plot using hue

```python
sns.relplot(data=tips,x='day',y='total_bill',kind='scatter',hue='smoker',style='sex')
plt.show()
```

### Task-2:Line Plot as scatter & facet

# 1.Line plot

```python
sns.lineplot(data=tips,x='day',y='total_bill',hue='smoker',style='time')
plt.show()
```

# 2.Scatter plot

```python
sns.scatterplot(data=tips,x='day',y='total_bill',hue='smoker',style='time')
plt.show()
```

# 3.Faceting(col)

```python
sns.relplot(data=tips,x='day',y='total_bill',col='sex',kind='line')
plt.show()
```

### Task-3:Distribution plot

# 1.Histogram plot()

```python
sns.histplot(data=tips,x='total_bill',bins=15,hue='sex',element='step')
plt.show()
```

# 2.Kernal Density Estimation(Kde plot)

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,hue='sex')
plt.show()
```

# 3.Rug Plot()

```python
sns.rugplot(data=tips,x='total_bill',hue='sex')
plt.show()
```

# 4.Hist + Kde Plot

```python
sns.histplot(data=tips,x='total_bill',bins=15,hue='sex',kde=True,element='step')
plt.show()
```

### Task-4:Bivariate Distribution plots
# 1.Bivariate Histogram plot()

```python
sns.histplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# 2.Bivariate kde plot()

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
plt.show()
```

### Task-5:Matrics Plot()
# 1.Pair plot()

```python
sales=pd.read_csv('Chocolate Sales (2).csv')
sales.head()
```

```python
sns.pairplot(data=sales)
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

# 2.Heatmap of correlation

```python
sales.head()
```

```python
sales['Amount'] = pd.to_numeric(sales['Amount'].str.replace(r'[\$,]', '', regex=True))
corr = sales[['Amount', 'Boxes Shipped']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```
